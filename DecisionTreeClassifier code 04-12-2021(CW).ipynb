#### Importing Libraries

```python
import pandas as pd
```

#### Loading Data

```python
ds=pd.read_csv('DT.csv')
ds.head(10)
```

Seperating Dependent and independent variables 
- Depending Variable: `salare>100k` (target)
- Independent Variables: `company` `job` `degree` (input1)

```python
input1 = ds.drop('salare>100k',axis='columns')
target = ds['salare>100k']
```

```python
target
```

```python
input1
```

>Since the dependent variables are in string type, (We cannot apply math on string types) We have to do label encoding,
i.e., we have to give numbers `0`,`1`,`2`,.. to the labels so that we can apply mathematical operations on it to train the model

```python
from sklearn.preprocessing import LabelEncoder 
comp=LabelEncoder()
job=LabelEncoder()
deg=LabelEncoder()
```

```python
input1['compn']=comp.fit_transform(input1['company'])
```

```python
input1['jobn']=job.fit_transform(input1['job'])
```

```python
input1['degn']=deg.fit_transform(input1['degree'])
```

Now we have seperate columns for each attribute/Variable, But we have to get rid of the original columns.(Why? Because we already have encoded them.)

```python
input1
```

Using `drop` method, we can drop the orginal columns!

```python
input2=input1.drop(columns=['company','job','degree'],axis=1)
input2
```

Now, Imorting `tree` module from `sklearn` library

Then creating a model using `DecisionTreeClassifier`, and using `fit` method to train it.

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(input2,target)
```

We have successfully trained our model, to check the accuracy, we can use `score` method

```python
score=model.score(input2,target)
score
```
